### 2.纯力学分析
**第一步**： 选取位置固定的锚链末端与锚的链接点作为参考点，建立力学分析模型，依次对浮标、钢桶、钢管和电焊锚链进行受力分析，稳态时，各物体受力平衡。  
**第二步**： 钢管、钢桶和电焊锚链的受力方向与物体间存在夹角，利用稳态时物体间的力矩平衡，对各夹角求解，当钢管、钢桶和电焊锚链长度在竖直距离的和为水深时，求出夹角和吃水深度便是所需值。  
**第三步**： 利用钢管、钢桶以及每节锚链的倾斜角度便可求出电焊锚链的形状和浮标的浮动范围

\begin{aligned}

\mathrm{T}_1 \cos \theta_1=\mathrm{F}_{\text {风 }} \\
\mathrm{T}_1 \sin \theta_1=\mathrm{F}_{\text {浮标 }}-\mathrm{G}_{\text {浮标 }}
\end{array}\right.
\end{aligned}
$$
$$
\mathrm{F}_{\text {风 }}=0.625 \mathrm{D}(2-\mathrm{h}) \mathrm{v}^2
$$

### 2.4总结方程
因此，可利用各个物体的长度$L_i$与各个物体与水平面的夹角$φ_i$之间的对应关系求得海水深度 H 与吃水深度 h 之间的关系为：$$\mathrm{H}=\sum_{i=1}^n L_i \sin \varphi_{\mathrm{i}}+\mathrm{h}$$   
也可利用各个物体的长度$L_i$与各个物体与水平面的夹角$φ_i$的对应关系求得浮标的游动半径为:$$\mathrm{R}=\sum_{i=1}^n L_i \cos \varphi_{\mathrm{i}}$$
将上述的力学分析和和力矩平衡的递推公式等进行整理，可得：
$$
\left\{\begin{array}{l}
\left\{\begin{array}{l}
\mathrm{T}_1 \cos \theta_1=0.625 \times \mathrm{D}(2-\mathrm{h}) \mathrm{v}^2 \\
\mathrm{~T}_1 \sin \theta_1=\rho_{\text {海水 }} g \pi r^2 h-\mathrm{m}_{\text {浮标 }} \mathrm{g}
\end{array}\right. \\
\left\{\begin{array}{l}
\left\{\mathrm{T}_{\mathrm{i}+1} \cos \theta_{\mathrm{i}+1}=\mathrm{T}_{\mathrm{i}} \cos \theta_{\mathrm{i}}\right. \\
\mathrm{T}_{\mathrm{i}+1} \sin \theta_{\mathrm{i}+1}=\mathrm{T}_{\mathrm{i}} \sin \theta_{\mathrm{i}}-\left(\mathrm{G}_{\mathrm{i}}-\mathrm{F}_{\mathrm{i}} \text { 浮 }\right)
\end{array} \quad(\mathrm{i} \neq 5)\right. \\
\left\{\begin{array}{l}
\mathrm{T}_6 \cos \theta_6=\mathrm{T}_5 \cos \theta_5 \\
\mathrm{~T}_6 \sin \theta_6=\mathrm{T}_5 \sin \theta_5-\left(\mathrm{G}_{\text {球 }}-\mathrm{F}_{\text {浮球 }}\right)-\left(\mathrm{G}_5-\mathrm{F}_{\text {浮 } 5}\right)
\end{array} \quad(\mathrm{i}=5)\right. \\
\varphi_{\mathrm{i}}=\arctan \left(\frac{2 \mathrm{~T}_{\mathrm{i}} \sin \theta_{\mathrm{i}}-\left(\mathrm{G}_{\mathrm{i}}-\mathrm{F}_{\text {浮 } \mathrm{i}}\right.}{2 \mathrm{~T}_{\mathrm{i}} \cos \theta_{\mathrm{i}}}\right),(\mathrm{i}=1,2, \cdots, \mathrm{n}) \\
\mathrm{H}=\sum_{i=1}^n L_i \sin \varphi_{\mathrm{i}}+\mathrm{h} \\
\mathrm{R}=\sum_{i=1}^n L_i \cos \varphi_{\mathrm{i}}
\end{array}\right.
$$

总体思路解析：不论是钢管、钢桶还是锚链，每一部分都蕴含3个未知量。浮标蕴含1个未知量。

钢管、钢桶、锚链有3个等式：受力平衡\*2，力矩平衡\*1；
末端锚链只有1个等式：力矩平衡\*1；
浮标只有2个等式：受力平衡\*2；
另有总高度为18的等式。
关于末端锚链为什么不考虑受力平衡：涉及到下方的拉力，这个不在我们的考虑范围内，不是我们的未知量，而且这个拉力的大小和方向是由系泊系统前面部分的状态所决定的，有因果性，可以计算出来，如果将其视为我们的未知量，就会多两个未知数和两个方程，同样也是方程数等于未知数个数。

```python
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd
import math

plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
dBuoy = 2 #浮标直径
lBuoy = 2 #浮标长度
mBuoy = 1000 #浮标质量

dTube = 0.05 #钢管直径
lTube = 1
mTube = 10

lBarrel = 1 #钢桶长度
dBarrel = 0.3
mBarrel = 100

sChain = 22.05 #锚链长度
lChainRing = 0.105 #链环长度
rhoChainRing = 7 #链环密度
mChainRing = lChainRing * rhoChainRing #链环质量

mBall = 1200 #重物球质量
H = 18 #海水深度
rhoWater = 1025 #海水密度
rhoSteel = 7850 #钢铁密度
g = 9.8 #重力加速度
```

```python
v = 12 #风速

#浮力
def fFloat(d,h):
    return rhoWater * g * math.pi * 0.25 * (d ** 2) * h

def fWind(h,v):
    return 0.625 * ((lBuoy - h) * dBuoy) * (v ** 2)

for i in range(0,20000):
    h = 0.0001 * i

    #水平分力
    Tx = fWind(h,v)

    #竖直分力
    Ty = [0]

    #物体与水平面的倾斜角
    phi = [0]

    #对浮标做受力分析，求出Ty[1]
    Ty.append(fFloat(dBuoy,h) - mBuoy * g)
    #print(Ty)
    phi.append(math.atan((2 * Ty[1] - (mTube * g - fFloat(dTube,lTube))) / 2 / Tx))

    #递归求得所有钢管连接处的Ty
    for i in range(2,6):
        Ty.append(Ty[i-1] - (mTube * g - fFloat(dTube,lTube)))
        phi.append(math.atan((2 * Ty[i] - (mTube * g - fFloat(dTube,lTube))) / 2 / Tx))

    #锚链与钢桶连接处的Ty
    vBall = mBall / rhoSteel
    fFloatBall = vBall * rhoWater * g
    Ty.append(Ty[5] - (mBarrel * g - fFloat(dBarrel,lBarrel)) - (mBall * g - fFloatBall))
    phi.append(math.atan((2 * Ty[i] - (mBarrel * g - fFloat(dBarrel,lBarrel))) / 2 / Tx))
    #print(Ty[6])

    #锚环连接处的Ty
    vChainRing = mChainRing / rhoSteel
    fFloatRing = vChainRing * rhoWater * g
    for i in range(7,216):
        Ty.append(Ty[i-1] - (mChainRing * g - fFloatRing))
        phi.append(math.atan((2 * Ty[i] - (mChainRing * g - fFloatRing)) / 2 / Tx))

    # print('Ty : ',Ty)
    # print('phi : ',phi)

    #每个节点的x坐标
    x = [0] * 217

    #每个节点的y坐标
    y = [0] * 217

    for i in range(215,5,-1):
        x[i] = x[i + 1] + lChainRing * math.cos(phi[i])
        y[i] = y[i + 1] + lChainRing * math.sin(phi[i])


    x[5] = x[6] + lBarrel * math.cos(phi[5])
    y[5] = y[6] + lBarrel * math.sin(phi[5])

    for i in range(4,0,-1):
        x[i] = x[i + 1] + lTube * math.cos(phi[i])
        y[i] = y[i + 1] + lTube * math.sin(phi[i])

    if (abs(y[1] - 18) < 0.1):
        break


plt.plot(x[1:-1],y[1:-1])
plt.show()
print('H = ',y[1])
print('R = ',x[1])
print('h = ',h)
#print(Ty[6])

for i in range(1,6):
    print('倾斜角度 = ',90 - phi[i] / math.pi * 180)

#print('Ty : ',Ty)
#print('phi : ',phi)
# print('x : ',x)
# print('y : ',y)
```

悬链线 (Catenary)指的是一种曲线，指两端固定的一条(粗细与质量分布)均匀、柔软(不能伸长)的链条，在重力的作用下所具有的曲线形状，例如悬索桥等，因其与两端固定的绳子在均匀引力作用下下垂相似而得名。适当选择坐标系后，悬链线的方程是一个双曲余弦函数，其标准方程为：y=a cosh(x/a)，其中，a为曲线顶点到横坐标轴的距离。

### 3.3.1浮标的静力学分析
有如下公式：
$$
\begin{gathered}
\sum F_x=0: \quad F_{\text {风 }}-F_1 \sin \theta_1=0 \\
\sum F_y=0: \quad F_{\text {浮 }}-G-F_1 \cos \theta_1=0 \\

\end{gathered}
$$  
第一问中考虑风速，浮标受到重力、浮力、钢管的拉力、风力四个力。在静力平衡状态下，四力构成平面汇交力系。

### 3.3.2钢桶和重物球的静力学分析
**重物球浮力分析**
  
上接第 4 节钢管的钢桶。钢桶下接电焊锚链悬挂重物球，使钢桶的倾斜角度尽可能小。下讨论重物球在海水中受到的浮力对实际情况的影响：
$$
\left\{\begin{array}{l}
F_{\text {浮 }}=\rho_{\text {海水 }} g V \\
m=\rho \cdot V \\
G=m g \\
V=\frac{4}{3} \pi R^2
\end{array} \quad \text { 联立左式, 解得: } \quad F_{\text {浮 }}=\frac{\rho_{\text {海水 }}}{\rho} \cdot G\right.
$$  
由上式可知：重物球所受浮力**不可忽略**！！   
  
**受力平衡**
$$
\begin{aligned}
&\sum F_x=0: F_5 \sin \theta_5-F_{\text {锚 }} \sin \theta_{\text {锚 }}=0 \\
&\sum F_y=0: F_5 \cos \theta_5+F_{\text {浮桶 }}+F_{\text {浮球 }}-G_{\text {桶 }}-G_{\text {球 }}-F_{\text {锚 }} \cos \theta_{\text {锚 }}=0
\end{aligned}
$$
由几何关系得: $H_{\text {桶 }}=\cos \alpha_{\text {桶 }} \cdot l$  
  
**力矩平衡**
$$
F_5 \times \frac{1}{2} \times \sin \left(\alpha_{\text {桶 }}-\theta_4\right)-F_{\text {锚 }} \times \frac{1}{2} \times \sin \left(\theta_{\text {锚 }}-\alpha_{\text {桶 }}\right)=0
$$

### 3.3.3锚链的静力学分析
设$ds$与海床的夹角为$\phi$，则由几何关系推得：
$$\sin \varphi=\frac{F_{\text {坚直 }}}{\sqrt{F_{\text {坚直 }}^2+F^2 \text { 水平 }}}$$
进而得出$$\mathrm{dy}=\mathrm{ds} \cdot \sin \varphi=d s \cdot \frac{F_{\text {坚直 }}}{\sqrt{F_{\text {坚直 }}^2+F_{\text {水平 }}^2}}$$
$$
\begin{aligned}
&F_{\text {水平 }}=F_{\text {锚 }} \sin \theta_{\text {锚}}=F_0 \sin \theta_0 \\
&F_{\text {骍直 }}=F_{\text {锚 }} \cos \theta_{\text {锚}}=m g s-\rho g A s+F_0 \cos \theta_0
\end{aligned}
$$  
对$dy$的表达式积分，有：
$$\mathrm{y}=\int_0^S d y=\int_0^S d s \cdot \sin \varphi=\int_0^S d s \cdot \frac{F_{\text {坚直 }}}{\sqrt{F_{\text {坚直 }}^2+F_{\text {水平 }}^2}}$$  
为简化计算，令
$$
\begin{gathered}
a=m g-\rho g A \\
b=F_0 \cos \theta_0 \\
c=F_0 \sin \theta_0
\end{gathered}
$$  

积分后有：
$$
y=\frac{\sqrt{(a s+b)^2}-\sqrt{b^2+c^2}}{a}
$$
$$
x=\frac{c}{a} \cdot \ln \left(\frac{a s+b+\sqrt{(a s+b)^2+c^2}}{b+\sqrt{b^2+c^2}}\right)
$$

```python
#浮力
def fFloat(d,h):
    return rhoWater * g * math.pi * 0.25 * (d ** 2) * h

def fWind(h,v):
    return 0.625 * ((lBuoy - h) * dBuoy) * (v ** 2)

v = 12

for i in range(0,20000): #其实入水深度不能太小，否则浮力太小
    h = i * 0.0001

    #水平分力
    Tx = fWind(h,v)

    #竖直分力
    Ty = [0]

    #物体与水平面的倾斜角
    phi = [0]

    #对浮标做受力分析，求出Ty[1]
    Ty.append(fFloat(dBuoy,h) - mBuoy * g)
    phi.append(math.atan((2 * Ty[1] - (mTube * g - fFloat(dTube,lTube))) / 2 / Tx))

    #递归求得所有钢管连接处的Ty
    for i in range(2,6):
        Ty.append(Ty[i-1] - (mTube * g - fFloat(dTube,lTube)))
        phi.append(math.atan((2 * Ty[i] - (mTube * g - fFloat(dTube,lTube))) / 2 / Tx))

    #锚链与钢桶连接处的Ty
    Ty.append(Ty[5] - (mBarrel * g - fFloat(dBarrel,lBarrel)) - (mBall * g - fFloatBall))
    phi.append(math.atan((2 * Ty[i] - (mBarrel * g - fFloat(dBarrel,lBarrel))) / 2 / Tx))

    #print(Ty)

    #悬链线
    A = vChainRing / lChainRing #锚环的横截面积
    a = rhoChainRing * g - rhoWater * g * A
    b = 0 #由于把锚链看成悬链线，它接触海床的地方切线斜率为0
    c = Tx
    #锚链悬在水中的实际长度
    realS = (fFloatBall + 4 * fFloat(dTube,lTube) + fFloat(dBarrel,lBarrel) + fFloat(dBuoy,h) - mBuoy * g - mBall * g - mBarrel * g - 4 * mTube * g) / (rhoChainRing - rhoWater * A) / g
    if realS < 0: #实际长度小于0，说明入水深度太小
        continue
    yChain = c / a * (math.sqrt((a * realS / c + b / c) ** 2 + 1) - math.sqrt((b / c) ** 2 + 1))
    #print(yChain)

    #每个节点的x坐标
    x = [0] * 7

    #每个节点的y坐标
    y = [0] * 7

    y[6] = yChain

    y[5] = y[6] + lBarrel * math.sin(phi[5])

    for i in range(4,0,-1):
        y[i] = y[i + 1] + lTube * math.sin(phi[i])

    if (abs(y[1] + h - 18) < 0.01):
        break

#print(y)
print('h = ',h)
print('phi : ',phi)
print('H = ',y[1] + h)
print('海床上的锚链长度 = ',sChain - realS)

for i in range(1,6):
    print('倾斜角度 = ',90 - phi[i] / math.pi * 180)

def yChain(s): #把cos（atan（x））化简后的结果
    return c / a * (math.sqrt((a * s / c + b / c) ** 2 + 1) - math.sqrt((b / c) ** 2 + 1))

def xChain(s): #
    return c / a * math.log((a * s + b + math.sqrt((a * s + b) ** 2 + c ** 2)) / (b + math.sqrt(b ** 2 + c ** 2)))

x = []
y = []

i = 0
while i * lChainRing < (sChain - realS):
    x.append(i * lChainRing)
    y.append(0)
    i += 1

s = 0.1
while i < 211:
    x.append(sChain - realS + xChain(s))
    y.append(yChain(s))
    i += 1
    s += 0.1

plt.plot(x,y)
plt.plot([sChain - realS],[0],'ro')
plt.show()
```

### 4.两种思路对比  
**相同**
面对复杂方程组，都采用遍历法求解；

都对模型做出合理且必须的假设；

都借助了力学中的受力分析和力矩平衡的思想建立变量之间的关系

**不同**
纯力学分析方法由于在锚链部分有迭代、因果的思想，采取隔离法逐一分析，所以对于锚链躺在海床上的情况不需单独讨论；
悬链线方法由于对锚链整体视为线性缆索，采用整体法分析，舍弃了对各组分复杂联动的考虑，以换取计算的简便——但也因此缺少了许多信息，所以对于锚链是否躺在海床上要分类讨论。
